## Гипотеза

Если добавить в десктопную версию вкладки для чатов, мы увеличим Time spend на 20%, потому что станет удобнее пользоваться приложением с ПК.

## Что делаем

Контрольная версия: оставляем текущий вид приложения. Тестовая версия: Добавляем в приложение новый функционал: возможность использовать вкладки для чатов.

## На каких пользователях тестируем:

Отбираем 5000 пользователей, которые пользуются приложением более 3 месяцев и проводят не меньше 1 часа в приложении. Делим на 2 группы:
группа А - контрольная;
группа B - тестовая.

## Метрики:

Time spend

## План действий

<b>Если</b> наш эксперимент будет положительным и мы зафиксируем ожидаемое улучшение в в ключевых метриках, то масштабируем изменение и «выкатываем» его на всех пользователей.

<b>Если</b> метрики растут недостаточно, откатываем эксперимент.
